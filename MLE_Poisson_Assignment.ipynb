## 1. Simulate 100 visits to your youtube channel, assuming that they will a Poisson distribution with a mean of 10 visits per minute. Plot the arrival time vs visitor index.

```python
from scipy.stats import poisson
from datascience import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plots
 

# Poisson Distribution
PoisDist = Table().with_column('Arrival Time Vs. Vistor ',np.random.poisson(lam=5,size=100))
PoisDist.hist()

# Log-Likelihood Function
def loglikeli(lam):
  logs = make_array()
  for k in PoisDist.column(0):
    logs = np.log(poisson.pmf(k=k,mu = lam))
  return np.sum(logs)

# lambdas
lams = np.arange(1,10,0.01)

# Likelihood
ls = make_array()
for lam in lams:
  print(lam)
  ls = np.append(ls, np.exp(loglikeli(lam)))


plots.plot(lams,ls)
```

Assume that you are given the visitation data generated in Step 1, implement a Gradient Descent algorithm from scratch that will estimate the Poisson distribution according to the Maximum Likelihood criterion. Plot the estimated mean vs iterations to showcase convergence towards the true mean.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (11,5)  #set figure size
import numpy as np
from numpy import exp
from scipy.special import factorial
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import statsmodels.api as sm
from scipy import stats
from scipy.stats import norm
from statsmodels.iolib.summary2 import summary_col
```

```python
poisson_pmf = lambda y,mu: mu**y /factorial(y) * exp(-mu)
y_values =range(0, 25)

fig,ax = plt.subplots(figsize =(12,10))

for mu in [1,4,10]:
  distribution =[]
  for y_i in y_values:
    distribution.append(poisson_pmf(y_i, mu))
  ax.plot(y_values, distribution)

ax.grid()
ax.set_xlabel('Estimate mean',fontsize =14)
ax.set_ylabel("Iteration", fontsize=14)
ax.axis(xmin=0, ymin=0)


plt.show()
```
