## Exercise 3
a. Calculate the probability of the counts of outcome “2” if the possible **equally probable** outcomes are all integer numbers between 1 and 10 and we performed n=1000 trials.

```
import numpy as np
trials = 1000

count = 0 #Setting count to zero

for i in range(trials):
  ints = np.random.randint(low=1, high=11, size=10) #Outcomes are all integer numbers between 1 and 10
  count += np.sum(ints==2)*1 #Counting number of times 2 occured #multiplying 1 to the vector it will be 0or1.

prob = count/(trials*10)
print("Probability of outcome as '2' from 1-10 with 1000 trials is: ", prob)
```

b. Simulate n independent trials of the multinoulli (categorical distribution) compliant to the specification of (a). Plot the probability in (a) as a function of n independent trials (n=10-1000). Write your conclusions with respect to the behavior of the estimated probability as nn increases.

```
import matplotlib.pyplot as plt

probList = [] #list to store probabilities
trailsList = [] #list to store the number of trails

for n in range(10, 1001):
  count = 0
  for i in range(n):
    ints = np.random.randint(low=1, high=11, size=10) #Outcomes are all integer numbers between 1 and 10
    count += np.sum(ints==2)*1 #Counting number of times 2 occured #multiplying 1 to the vector it will be 0or1.

  prob = count/(10*n)

  #storing the probability in the probList
  probList.append(prob)
  trailsList.append(n) 

#plotting on the results on graph
plt.plot(trailsList,probList)
plt.xlabel('Trials')
plt.ylabel('Probability')
plt.title("Probability of 2")
plt.show()
```

## Exercise 4

Correlated Random Variables Bivariate Gaussian Distribution **Figure**.

```
import numpy as np
import matplotlib.pyplot as plt


n=1000  # no of sample

#------------------plot 1----------------------------------------
p=0.0 

mean = [0, 0]
cov = [[1, p], [p, 1]]
x, y = np.random.multivariate_normal(mean, cov, n).T

plt.subplot(2, 2, 1)
plt.plot(x, y, '.',color= 'black')
plt.ylabel('X2')
plt.title("p=0.00")
plt.axis('equal')

#------------------plot 2----------------------------------------
p=0.20

mean = [0, 0]
cov = [[1, p], [p, 1]]
x, y = np.random.multivariate_normal(mean, cov, n).T

plt.subplot(2, 2, 2)
plt.plot(x, y, '.',color= 'black')
plt.title("p=0.20")
plt.axis('equal')

print("")

#------------------plot 3----------------------------------------
p=-0.80  

mean = [0, 0]
cov = [[1, p], [p, 1]]
x, y = np.random.multivariate_normal(mean, cov, n).T

plt.subplot(2, 2, 3)
plt.plot(x, y, '.',color= 'black')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title("p=-0.80")
plt.axis('equal')


#------------------plot 4----------------------------------------
p=0.99

mean = [0, 0]
cov = [[1, p], [p, 1]]
x, y = np.random.multivariate_normal(mean, cov, n).T

plt.subplot(2, 2, 4)
plt.plot(x, y, '.',color= 'black')
plt.xlabel('X1')
plt.title("p=0.99")



plt.tight_layout()
plt.show()
```
